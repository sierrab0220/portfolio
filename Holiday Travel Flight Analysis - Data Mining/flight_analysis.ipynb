````python
def download_and_process(url, zip_filename, csv_filename, verbose=True, dtypes=None):
    """
    Complete data pipeline for processing BTS flight data files.

    Args:
        url (str): Source URL for the zip file
        zip_filename (str): Name for saving the zip file
        csv_filename (str): Name for the output CSV file
        verbose (bool): Whether to print detailed logs (default: True)
        dtypes (dict): Optional dictionary of column names and data types for memory optimization.

    Returns:
        bool: Success status of the operation.
    """
    try:
        if verbose:
            print(f"üì• Downloading {zip_filename}...")

        # Set up file paths
        zip_path = os.path.join(ZIP_FOLDER, zip_filename)
        csv_path = os.path.join(PROJECT_FOLDER, csv_filename)

        # Download and save zip file
        response = requests.get(url, stream=True)
        response.raise_for_status()

        with open(zip_path, 'wb') as f:
            f.write(response.content)
        if verbose:
            print(f"üíæ Saved zip file to: {zip_path}")

        # Process the zip file
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            # Locate ASC file
            asc_files = [f for f in zip_ref.namelist() if f.endswith('.asc')]

            if not asc_files:
                print("‚ùå No .asc file found in the zip file")
                return False

            # Process ASC file
            asc_file = asc_files[0]
            if verbose:
                print(f"üîÑ Found ASC file: {asc_file}")

            # Convert to DataFrame
            asc_content = zip_ref.read(asc_file)
            df = pd.read_csv(
                io.StringIO(asc_content.decode('utf-8')),
                delimiter='|',
                low_memory=False,
                dtype=dtypes  # Apply optional memory optimization
            )

            # Ensure proper datetime conversion for date and time columns
            datetime_columns = ['flight_date']
            time_columns = ['scheduled_departure_time', 'actual_departure_time',
                            'scheduled_arrival_time', 'actual_arrival_time']

            # Convert date column to datetime
            for col in datetime_columns:
                if col in df.columns:
                    try:
                        df[col] = pd.to_datetime(df[col], errors='coerce', format='%Y%m%d')
                        if verbose:
                            print(f"üìÖ Converted column {col} to datetime format.")
                    except Exception as e:
                        print(f"‚ö†Ô∏è Error converting column {col} to datetime: {e}")

            # Convert time columns to datetime.time
            from datetime import datetime
            for col in time_columns:
                if col in df.columns:
                    try:
                        df[col] = df[col].apply(
                            lambda x: datetime.strptime(str(x), '%H:%M:%S').time()
                            if pd.notnull(x) else None
                        )
                        if verbose:
                            print(f"‚è∞ Converted column {col} to time format.")
                    except Exception as e:
                        print(f"‚ö†Ô∏è Error converting column {col} to time: {e}")

            # Optimize data types for classification models
            if 'destination_airport' in df.columns:
                df['destination_airport'] = df['destination_airport'].astype('category')
            if 'departure_delay' in df.columns:
                df['departure_delay'] = pd.to_numeric(df['departure_delay'], errors='coerce').astype('int32')
            if 'is_delayed' not in df.columns and 'departure_delay' in df.columns:
                df['is_delayed'] = (df['departure_delay'] >= 15).astype('int8')

            # Save processed data
            df.to_csv(csv_path, index=False)
            if verbose:
                print(f"‚úÖ Successfully saved to: {csv_path}")
                print(f"üìä Processed rows: {len(df):,}")

            return True

    except requests.exceptions.RequestException as req_err:
        print(f"‚ùå Network error while downloading {zip_filename}: {str(req_err)}")
    except zipfile.BadZipFile:
        print(f"‚ùå Invalid ZIP file: {zip_filename}")
    except Exception as e:
        print(f"‚ùå Unexpected error processing {zip_filename}: {str(e)}")
    return False




# üï° Define data sources with clear naming
data_sources = {
    ("November22_FlightData.zip", "November22_FlightData.csv"):
        "https://www.bts.gov/sites/bts.dot.gov/files/docs/legacy/additional-attachment-files/ONTIME.TD.202211.REL01.10JAN2023.zip",
    ("December22_FlightData.zip", "December22_FlightData.csv"):
        "https://www.bts.gov/sites/bts.dot.gov/files/docs/legacy/additional-attachment-files/ONTIME.TD.202212.REL03.10MAR2023.zip",
    ("January23_FlightData.zip", "January23_FlightData.csv"):
        "https://www.bts.gov/sites/bts.dot.gov/files/docs/legacy/additional-attachment-files/ONTIME.TD.202301.REL02.16MAR2023.zip"
}

# Process all files
for (zip_name, csv_name), url in data_sources.items():
    print(f"\nüîÑ Processing {zip_name}")
    print("=" * 50)
    success = download_and_process(url, zip_name, csv_name)
    if success:
        print(f"‚úÖ Successfully processed {zip_name} to {csv_name}")
    else:
        print(f"‚ùå Failed to process {zip_name}")
    print("=" * 50)

def cleanup_zip_files():
    """
    Verifies CSV existence and removes ZIP files to save space

    Returns:
        float: Total space saved in MB
    """
    space_saved = 0

    # First verify CSV files exist
    csv_files_exist = all(
        os.path.exists(os.path.join(PROJECT_FOLDER, f"{month}_FlightData.csv"))
        for month in ["November22", "December22", "January23"]
    )

    if not csv_files_exist:
        print("‚ùå Not all CSV files found. Keeping ZIP files as backup.")
        return 0

    print("\nüîç Beginning cleanup process...")

    # Remove ZIP files and track space saved
    for file in os.listdir(ZIP_FOLDER):
        if file.endswith('.zip'):
            file_path = os.path.join(ZIP_FOLDER, file)
            # Calculate file size before deletion
            size_mb = os.path.getsize(file_path) / (1024 * 1024)
            space_saved += size_mb

            # Remove the file
            os.remove(file_path)
            print(f"üóëÔ∏è Removed: {file} ({size_mb:.2f} MB)")

    print("\n‚ú® Cleanup Complete!")
    print(f"üíæ Total space saved: {space_saved:.2f} MB")

    return space_saved

# Run cleanup
total_saved = cleanup_zip_files()

def read_and_format_flight_data(csv_path):
    """
    Reads and formats flight data from CSV into clean DataFrame

    Args:
        csv_path (str): Path to the CSV file

    Returns:
        pandas.DataFrame: Cleaned and formatted flight data
    """
    # üüß Define our column mapping
    column_names = {
        8: 'flight_date',
        6: 'origin_airport',
        7: 'destination_airport',
        10: 'scheduled_departure_time',
        12: 'actual_departure_time',
        20: 'departure_delay',
        13: 'scheduled_arrival_time',
        15: 'actual_arrival_time'
    }

    print(f"\nüîÑ Processing: {os.path.basename(csv_path)}")

    # üì• Read the CSV
    df = pd.read_csv(csv_path)
    print(f"üìä Initial columns: {len(df.columns)}")

    # üî® Create cleaned DataFrame
    formatted_df = pd.DataFrame()

    # üüß Map and transform each column
    for col_index, new_name in column_names.items():
        formatted_df[new_name] = df.iloc[:, col_index]

    # üìÖ Convert date to datetime
    formatted_df['flight_date'] = pd.to_datetime(formatted_df['flight_date'], format='%Y%m%d')

    # üöπ Clean time fields
    time_columns = ['scheduled_departure_time', 'actual_departure_time',
                   'scheduled_arrival_time', 'actual_arrival_time']
    for col in time_columns:
        formatted_df[col] = formatted_df[col].astype(str).str.split('.').str[0]

    # üî¢ Convert delays to numeric
    formatted_df['departure_delay'] = pd.to_numeric(formatted_df['departure_delay'], errors='coerce')

    print(f"‚úÖ Processed columns: {len(formatted_df.columns)}")
    return formatted_df

# üìã List of CSV files to process
csv_files = [
    "November22_FlightData.csv",
    "December22_FlightData.csv",
    "January23_FlightData.csv"
]

# üìä Create empty list for DataFrames
dfs = []

# üîÑ Process each CSV file
for csv_file in csv_files:
    file_path = os.path.join(PROJECT_FOLDER, csv_file)
    if os.path.exists(file_path):
        df = read_and_format_flight_data(file_path)
        dfs.append(df)
        print(f"‚úÖ Processed {csv_file}: {len(df):,} rows")
    else:
        print(f"‚ùå File not found: {csv_file}")

# üîÑ Combine all DataFrames
if dfs:
    combined_df = pd.concat(dfs, ignore_index=True)
    print(f"\nüìä Total rows in combined dataset: {len(combined_df):,}")
    print("\nüëÄ First few rows of combined data:")
    print(combined_df.head().to_markdown())
else:
    print("‚ùå No data frames to combine")


def create_mia_flights(df):
    """
    Creates and cleans a DataFrame containing only flights originating from MIA.
    Optimized data types for classification with is_delayed as target.
    """
    # Filter for MIA departures
    mia_flights = df[df['origin_airport'] == 'MIA'].copy()

    # Drop the origin_airport column since it's now redundant
    mia_flights = mia_flights.drop('origin_airport', axis=1)

    # Reset index to ensure clean sequential indexing
    mia_flights = mia_flights.reset_index(drop=True)

    # Convert flight_date to datetime64[ns]
    #mia_flights['flight_date'] = pd.to_datetime(mia_flights['flight_date'])

    # Define time columns
    time_columns = ['scheduled_departure_time', 'actual_departure_time',
                    'scheduled_arrival_time', 'actual_arrival_time']

    # Function to standardize time format and handle edge cases
    def standardize_time(time_str):
        from datetime import datetime, time
        # Clean the string and ensure 4 digits
        time_str = str(time_str).replace(':00', '').replace(':', '')
        time_str = time_str.zfill(4) if time_str not in ['0', '', 'nan'] else '0000'

        # Handle 24:00 case
        if time_str.startswith('24'):
            time_str = '00' + time_str[2:]

        # Convert to datetime.time object
        hour = int(time_str[:2])
        minute = int(time_str[2:])
        return time(hour, minute)

    for col in time_columns:
        mia_flights[col] = mia_flights[col].apply(standardize_time)

    # Convert destination_airport to category (memory efficient, ready for encoding)
    mia_flights['destination_airport'] = mia_flights['destination_airport'].astype('category')

    # Convert numeric columns with appropriate types
    mia_flights['departure_delay'] = mia_flights['departure_delay'].astype('int32')
    mia_flights['is_delayed'] = (mia_flights['departure_delay'] >= 15).astype('int8')

    # Print single summary with data types and sample data
    print("\n=== MIA Flights Dataset Summary ===")
    print(f"‚úàÔ∏è  Total Flights: {len(mia_flights):,}")
    print(f"üìÖ  Date Range: {min(mia_flights['flight_date'])} to {max(mia_flights['flight_date'])}")
    print(f"üö®  Delayed Flights (15+ min): {mia_flights['is_delayed'].sum():,} ({mia_flights['is_delayed'].mean()*100:.1f}%)")
    print(f"üõ™Ô∏è  Top Destinations: {', '.join(mia_flights['destination_airport'].value_counts().head().index)}")

    print("\nüìä Data Types:")
    print(mia_flights.info())

    print("\nüìã Sample Data:")
    print(mia_flights.head().to_markdown(index=False))

    # Save to CSV in Google Drive
    output_path = "/content/drive/MyDrive/Flight_Delay_Project/mia_flights_processed.csv"
    mia_flights.to_csv(output_path, index=False)
    print(f"\n‚úÖ DataFrame saved to: {output_path}")

    return mia_flights



# Call the function
mia_flights = create_mia_flights(combined_df)

mia_flights
````

````python
def suggest_outlier_method(df):
    methods = {}

    for col in merged_df.select_dtypes(include=np.number).columns:
        data = merged_df[col].dropna()

        # Skip columns with very few unique values (like binary data)
        if data.nunique() < 10:
            methods[col] = "Other (Likely categorical)"
            continue

        # Calculate skewness
        skewness = data.skew()

        # Calculate IQR for Tukey method
        Q1 = np.percentile(data, 25)
        Q3 = np.percentile(data, 75)
        IQR = Q3 - Q1

        # Calculate outliers using Tukey method
        lower_lim = Q1 - 1.5 * IQR
        upper_lim = Q3 + 1.5 * IQR
        tukey_outliers = data[(data < lower_lim) | (data > upper_lim)]

        # Log transformation candidates (high skewness)
        if abs(skewness) > 1:
            methods[col] = "Log transformation (highly skewed)"

        # Winsorize candidates (data with a few extreme outliers)
        elif len(tukey_outliers) / len(data) > 0.01:
            methods[col] = "Winsorize (outliers present)"

        # Tukey method (moderate skewness and outliers)
        else:
            methods[col] = "Tukey (mild/moderate outliers)"

    return methods

# Apply the function and print the results
outlier_methods = suggest_outlier_method(df)
for column, method in outlier_methods.items():
    print(f"Column: {column} - Suggested Method: {method}")
````

````python
import numpy as np
from scipy.stats import mstats

def apply_outlier_methods(df, methods, skip_columns=None):
    """
    Applies the suggested outlier methods (log transformation, winsorization, Tukey) to the specified columns,
    while skipping specified columns.

    Parameters:
        df (pd.DataFrame): The input DataFrame.
        methods (dict): Dictionary of column names and suggested methods.
        skip_columns (list): List of columns to skip from outlier handling.

    Returns:
        pd.DataFrame: The DataFrame with applied outlier handling.
    """
    if skip_columns is None:
        skip_columns = []  # Default to an empty list if none provided

    for column, method in methods.items():
        if column in skip_columns:
            print(f"Skipping outlier handling for column: {column}")
            continue  # Skip the specified columns

        if method == "Log transformation (highly skewed)":
            # Apply log transformation
            df[column] = np.log1p(df[column])  # Use log1p to handle zero values gracefully
        elif method == "Winsorize (outliers present)":
            # Winsorize the data to handle extreme outliers
            df[column] = mstats.winsorize(df[column], limits=(0.01, 0.01))  # Winsorize the top and bottom 1%
        elif method == "Tukey (mild/moderate outliers)":
            # Apply Tukey's method to remove mild/moderate outliers
            Q1 = np.percentile(df[column].dropna(), 25)
            Q3 = np.percentile(df[column].dropna(), 75)
            IQR = Q3 - Q1
            lower_lim = Q1 - 1.5 * IQR
            upper_lim = Q3 + 1.5 * IQR
            df = df[(df[column] >= lower_lim) & (df[column] <= upper_lim)]
        else:
            # Skip columns with "Other (Likely categorical)" or no method
            continue

    return df

# Example usage
outlier_methods = {
    "departure_delay": "Log transformation (highly skewed)",
    "is_delayed": "Other (Likely categorical)",
    "avg_wind_speed": "Tukey (mild/moderate outliers)",
    "precipitation": "Log transformation (highly skewed)",
    "max_temperature": "Log transformation (highly skewed)",
    "min_temperature": "Log transformation (highly skewed)",
    "some_other_column": "Other (Likely categorical)"  # Example for ignored columns
}

# Skip specific columns (e.g., "departure_delay")
columns_to_skip = ["departure_delay"]

# Apply outlier handling methods to the DataFrame
cleaned_df = apply_outlier_methods(merged_df, outlier_methods, skip_columns=columns_to_skip)

# View the cleaned DataFrame
print(cleaned_df.info())
````

### a. Feature Engineering

````python
# Create Feature Functions
def create_time_features(df):
    """
    Adds time-based cyclical features for hour and day of the week.
    """
    df = df.copy()
    if 'scheduled_departure_time' in df.columns and df['scheduled_departure_time'].dtype != 'datetime64[ns]':
        df['scheduled_departure_time'] = pd.to_datetime(df['scheduled_departure_time'])

    if 'flight_date' in df.columns and df['flight_date'].dtype != 'datetime64[ns]':
        df['flight_date'] = pd.to_datetime(df['flight_date'])

    df['hour'] = df['scheduled_departure_time'].dt.hour
    df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
    df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
    df['day_of_week'] = df['flight_date'].dt.dayofweek
    df['day_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
    df['day_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)
    return df

def create_weather_features(df):
    """
    Adds weather-related risk and temperature features.
    """
    df = df.copy()
    # Handle missing values
    weather_cols = ['avg_wind_speed', 'precipitation', 'max_temperature', 'min_temperature', 'fog', 'thunder']
    for col in weather_cols:
        if col in df.columns:
            df[col].fillna(0, inplace=True)

    # Avoid division by zero
    max_wind_speed = df['avg_wind_speed'].max() or 1
    max_precipitation = df['precipitation'].max() or 1

    df['weather_risk'] = (
        (df['avg_wind_speed'] / max_wind_speed) * 0.3 +
        (df['precipitation'] / max_precipitation) * 0.3 +
        df['fog'] * 0.2 +
        df['thunder'] * 0.2
    )
    df['temp_range'] = df['max_temperature'] - df['min_temperature']
    df['temp_mean'] = (df['max_temperature'] + df['min_temperature']) / 2
    return df

def create_flight_features(df):
    """
    Adds flight-specific features like buffer time and delay rates.
    """
    df = df.copy()
    if 'scheduled_arrival_time' in df.columns and 'scheduled_departure_time' in df.columns:
        df['buffer_time'] = (
            df['scheduled_arrival_time'] - df['scheduled_departure_time']
        ).dt.total_seconds() / 60
    else:
        df['buffer_time'] = 0  # Default if columns are missing

    df['dest_delay_rate'] = df.groupby('destination_airport')['is_delayed'].transform('mean')
    df['dest_flight_count'] = df.groupby('destination_airport')['flight_date'].transform('count')
    return df

# Main Pipeline with Info Display
def engineer_features(df):
    """
    Full pipeline for feature engineering: creates time, weather, and flight features.
    """
    print("Starting Feature Engineering Pipeline...")

    # Create features
    df = create_time_features(df)
    df = create_weather_features(df)
    df = create_flight_features(df)

    # Select features
    numeric_features = [
        'hour_sin', 'hour_cos', 'day_sin', 'day_cos',
        'weather_risk', 'temp_range', 'temp_mean',
        'avg_wind_speed', 'precipitation',
        'buffer_time', 'dest_delay_rate',
        'fog', 'thunder'
    ]

    categorical_features = ['destination_airport']
    features = numeric_features + categorical_features
    final_features = df[features]

    # Display info
    print("\nFeature Engineering Summary:")
    print("\nShape:", final_features.shape)
    print("\nNumeric Features Statistics:")
    print(final_features[numeric_features].describe())
    print("\nFeature Info:")
    print(final_features.info())

    return final_features

# Run Pipeline
X = engineer_features(cleaned_df)
````

````python
# Scoring metrics:
# 'accuracy', 'balanced_accuracy', 'average_precision', 'neg_brier_score', 'f1', 'f1_micro', 'f1_macro', 'f1_weighted',
# 'neg_log_loss', 'precision', 'recall', 'jaccard', 'roc_auc', 'roc_auc_ovr', 'roc_auc_ovo', 'roc_auc_ovr_weighted', 'roc_auc_ovo_weighted'

results = []
names = []
scoring = 'accuracy'    # Accuracy ONLY matters iff - the data is balanced!

for name, model in models:
    start = time()
    kfold = KFold(n_splits=10, random_state=7, shuffle=True)

    # Convert sparse matrix to dense array if necessary
    if name in ['LDA', 'QDA', 'NB', 'BNB']:  # Add other models that require dense input here
        X_train_balanced_dense = X_train_balanced.toarray()
        model.fit(X_train_balanced_dense, y_train_balanced)
        cv_results = cross_val_score(model, X_train_balanced_dense, y_train_balanced, cv=kfold, scoring=scoring)
    else:
        model.fit(X_train_balanced, y_train_balanced)
        cv_results = cross_val_score(model, X_train_balanced, y_train_balanced, cv=kfold, scoring=scoring)

    train_time = time() - start
    predict_time = time() - start
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
    print("Score for each of the 10 K-fold tests: ", cv_results)
    print(model)
    print("\tTraining time: %0.3fs" % train_time)
    print("\tPrediction time: %0.3fs" % predict_time)
    print()
````

### h. Model Calibration and Evaluation



````python
# =========================================
# Complete Model Calibration and Evaluation
# =========================================
def predict_in_batches(model, X):
    """
    Make predictions in batches, handling both dense and sparse matrices.
    """
    BATCH_SIZE = 10000
    n_samples = X.shape[0]
    predictions = []

    for i in range(0, n_samples, BATCH_SIZE):
        end_idx = min(i + BATCH_SIZE, n_samples)
        if scipy.sparse.issparse(X):
            batch = X[i:end_idx, :]
        else:
            batch = X[i:end_idx]
        pred = model.predict_proba(batch)[:, 1]
        predictions.append(pred)

    return np.concatenate(predictions)

# Model Calibration
print("\nStarting model calibration...")
calibrated_model = CalibratedClassifierCV(
    extratrees_model,
    cv=3,
    method='sigmoid',
    n_jobs=-1
)

calibrated_model.fit(X_train_balanced, y_train_balanced)
print("Calibration complete!")

# Generate Predictions
print("\nGenerating predictions...")
calibrated_probs = predict_in_batches(calibrated_model, X_test_processed)
uncalibrated_probs = predict_in_batches(extratrees_model, X_test_processed)
print("Predictions complete!")

# Calculate Performance Curves
print("\nCalculating performance curves...")
# Calibration curves
prob_true_uncal, prob_pred_uncal = calibration_curve(
    y_test, uncalibrated_probs, n_bins=10, strategy='quantile'
)
prob_true_cal, prob_pred_cal = calibration_curve(
    y_test, calibrated_probs, n_bins=10, strategy='quantile'
)

# ROC curve
fpr, tpr, _ = roc_curve(y_test, calibrated_probs)
roc_auc = roc_auc_score(y_test, calibrated_probs)

# Precision-Recall curve
precision, recall, _ = precision_recall_curve(y_test, calibrated_probs)
pr_auc = average_precision_score(y_test, calibrated_probs)
print("Curve calculations complete!")

# Plotting
print("\nGenerating plots...")
plt.figure(figsize=(15, 5))

# Calibration Curve
plt.subplot(131)
plt.plot(prob_pred_uncal, prob_true_uncal, 's-', label='Uncalibrated', color='#1f77b4')
plt.plot(prob_pred_cal, prob_true_cal, 'o-', label='Calibrated', color='#ff7f0e')
plt.plot([0, 1], [0, 1], '--', color='gray', label='Perfect calibration')
plt.xlabel('Mean predicted probability')
plt.ylabel('True probability')
plt.title('Calibration Curve')
plt.legend()
plt.grid(True, alpha=0.3)

# ROC Curve
plt.subplot(132)
plt.plot(fpr, tpr, color='#2ca02c', label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], '--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True, alpha=0.3)

# Precision-Recall Curve
plt.subplot(133)
plt.plot(recall, precision, color='#d62728', label=f'PR curve (AUC = {pr_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
print("Plotting complete!")

# Model Persistence
print("\nSaving model and metrics...")
model_info = {
    'model_name': 'ExtraTreesClassifier',
    'training_date': datetime.now().strftime('%Y-%m-%d'),
    'model_version': '1.0',
    'feature_names': list(X.columns),
    'performance_metrics': {
        'accuracy': accuracy_score(y_test, y_pred),
        'roc_auc': roc_auc,
        'pr_auc': pr_auc,
        'calibration_metrics': {
            'expected_calibration_error': np.mean(np.abs(prob_true_cal - prob_pred_cal)),
            'maximum_calibration_error': np.max(np.abs(prob_true_cal - prob_pred_cal))
        }
    }
}

# Save model and metrics
try:
    joblib.dump(calibrated_model, f"{PROJECT_FOLDER}/flight_delay_model.joblib")
    joblib.dump(model_info, f"{PROJECT_FOLDER}/model_metrics.joblib")
    print("Model and metrics saved successfully!")

    # Save performance metrics to a readable format
    with open(f"{PROJECT_FOLDER}/model_performance_log.txt", 'w') as f:
        f.write("Flight Delay Prediction Model Performance Log\n")
        f.write("==========================================\n")
        f.write(f"Model Version: {model_info['model_version']}\n")
        f.write(f"Training Date: {model_info['training_date']}\n")
        f.write("\nPerformance Metrics:\n")
        for metric, value in model_info['performance_metrics'].items():
            if isinstance(value, dict):
                f.write(f"\n{metric}:\n")
                for submetric, subvalue in value.items():
                    f.write(f"  {submetric}: {subvalue:.4f}\n")
            else:
                f.write(f"{metric}: {value:.4f}\n")
    print("Performance log saved!")

except Exception as e:
    print(f"Error saving model and metrics: {str(e)}")

print("\nModel evaluation and persistence complete!")
````
